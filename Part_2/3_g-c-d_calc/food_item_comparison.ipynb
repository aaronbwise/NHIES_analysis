```python
from pathlib import Path

import numpy as np
import pandas as pd
pd.set_option("display.max_rows", 150)
pd.set_option("display.max_columns", None)
```

```python
# Read old version
fn_old = 'table3_1.csv'
path_old = Path.cwd().joinpath('archive', fn_old)
df_old = pd.read_csv(path_old, skiprows=1)

# Read new version
fn_new = 'table_3_1_2020.csv'
path_new = Path.cwd().joinpath('data', fn_new)
df_new = pd.read_csv(path_new, skiprows=1)
```

```python
# Tidy columns
def tidy_columns(df):
    # Rename first column
    df = df.rename(columns={'Unnamed: 0': 'Food_Item'})

    # Clean whitespace from Food_Item
    df.loc[:, 'Food_Item'] = df.Food_Item.str.strip()

    # Create separate values for Food_Item column
    df['COICOP'] = [x.split(" ", 1)[0][:] for x in df.Food_Item]
    df['Label'] = [x.split(" ", 1)[1][:] for x in df.Food_Item]

    # Strip whitespace
    df.loc[:, 'COICOP'] = [x.strip() for x in df['COICOP']]
    df.loc[:, 'Label'] = [x.strip() for x in df['Label']]

    return df
```

```python
# Drop row if all values missing & clean column headings
df_old = tidy_columns(df_old.dropna(thresh=5))
df_new = tidy_columns(df_new.dropna(thresh=5))
```

```python
old_list = set(df_old.Food_Item.tolist())
new_list = set(df_new.Food_Item.tolist())
```
