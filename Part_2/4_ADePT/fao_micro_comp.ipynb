## Comparison of FAP micro ADePT output

```python
import os
import numpy as np
import pandas as pd
```

```python
# Read in food items dfs
micro_code_comp = pd.read_csv('micro_code_comp.csv')
fao_macro_code = pd.read_csv('fao_macro_code.csv')
fao_macro_upload = pd.read_csv('fao_macro_upload.csv')
```

```python
# Identify where there are differences
fao_micro_code_list = micro_code_comp['fao_code'].tolist()
unicef_micro_code_list = micro_code_comp['unicef_code'].tolist()

fao_macro_code_list = fao_macro_code['fao_macro_code'].tolist()
```

```python
set_a = set(fao_macro_code_list) & set(fao_micro_code_list) & set(unicef_micro_code_list)
set_x = set(fao_micro_code_list) - set_a
set_y = set(fao_macro_code_list) - set_a
set_z = set(unicef_micro_code_list) - set_a
# remove nan
set_z = {x for x in set_z if x==x}
```

```python
set_b = set_x & set_z
set_c = set_x & set_y
set_d = set_y & set_z
```

```python
set_e = set_z - (set_b | set_d)
set_f = set_x - (set_b | set_c)
set_g = set_y - (set_c | set_d)
```
